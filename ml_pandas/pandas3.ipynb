```python
lin_dat = pd.read_csv('insurance.csv')
lin_dat.head()
```

```python
print('how many null rows in dataframe')
lin_dat['age'][:3] = np.nan
lin_dat.loc[0][1]
```

```python
lin_dat.isna()
df4 = lin_dat.copy()
df4.dropna(axis=0,inplace=True,how='any')
df4.head()
```

```python
mean = lin_dat['age'].mean()
lin_dat.fillna(mean,inplace=True)
```

```python
duplicate = lin_dat[['age','sex']][:4]
duplicate2 = lin_dat[['age','sex']][5:6]
lin_dat1 = pd.concat([duplicate,duplicate2])
lin_dat1.head()
```

```python
lin_dat1.drop_duplicates(inplace=True)
print(lin_dat1.shape)
lin_dat1.head()
```
